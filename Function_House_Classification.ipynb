```python
def House_Classification(test):
  import matplotlib.pyplot as plt
  import numpy as np
  import cv2
  from ultralytics import YOLO
  model = YOLO('/content/best.pt')  # upload best file from Mydrive under House
  results = model(test)  # predict on an image
  names_dict = results[0].names

  probs = results[0].probs.data.tolist()


  image_path = test
  image = cv2.imread(image_path)

  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plot the image
  plt.imshow(image)
  plt.axis('off')  # Turn off axis
  plt.show()

  return  plt.show(), print(names_dict), print(probs), print(names_dict[np.argmax(probs)])
```
